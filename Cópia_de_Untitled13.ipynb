```python
import pandas as pd

# Carregar o arquivo CSV
df = pd.read_excel('df_final.xlsx')
```

```python
#Resposta da Questão 1

# Calcular a média
mean_price = df['priceEach'].mean()

# Calcular a mediana
median_price = df['priceEach'].median()

# Calcular a moda
mode_price = df['priceEach'].mode()[0]

print("Média do preço por unidade:", mean_price)
print("Mediana do preço por unidade:", median_price)
print("Moda do preço por unidade:", mode_price)
```

```python
#resposta da questão 2

import pandas as pd
import matplotlib.pyplot as plt

# Criar o histograma
plt.hist(df['amount'], bins=30, edgecolor='black')

# Adicionar título e rótulos
plt.title('Distribuição dos Valores Totais dos Pedidos')
plt.xlabel('Valor Total dos Pedidos')
plt.ylabel('Frequência')

# Mostrar o histograma
plt.show()
```

```python
# Resposta da questão 3

# Calcular o lucro por unidade vendida
df['profitPerUnit'] = df['priceEach'] - df['buyPrice']

# Calcular o primeiro quartil (Q1)
Q1 = df['profitPerUnit'].quantile(0.25)

# Calcular o segundo quartil (Q2 ou mediana)
Q2 = df['profitPerUnit'].quantile(0.50)

# Calcular o terceiro quartil (Q3)
Q3 = df['profitPerUnit'].quantile(0.75)

# Calcular a amplitude interquartílica (IQR)
IQR = Q3 - Q1

print("Primeiro quartil (Q1):", Q1)
print("Segundo quartil (Q2 ou mediana):", Q2)
print("Terceiro quartil (Q3):", Q3)
print("Amplitude interquartílica (IQR):", IQR)
```

```python
# Resposta da questão 4

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Criar o gráfico de caixa (boxplot)
plt.figure(figsize=(12, 8))
sns.boxplot(x='productLine', y='priceEach', data=df)

# Adicionar título e rótulos
plt.title('Distribuição do Preço por Unidade entre as Diferentes Linhas de Produto')
plt.xlabel('Linha de Produto')
plt.ylabel('Preço por Unidade')

# Girar os rótulos do eixo x para melhor legibilidade
plt.xticks(rotation=45)

# Mostrar o boxplot
```

```python
# Resposta da questão 5

# Criar o gráfico de dispersão
plt.figure(figsize=(10, 6))
sns.scatterplot(x='quantityOrdered', y='priceEach', data=df)
plt.title('Relação entre Quantidade Vendida e Preço por Unidade')
plt.xlabel('Quantidade Vendida')
plt.ylabel('Preço por Unidade')
plt.show()
```
